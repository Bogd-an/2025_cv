Розробити дві програми, які виконують такі дії з цифровим відео (ЦВ):
1. Перша програма, яка записується у контролер поворотних платформ, керує поворотною платформою з
двома серво приводами. Ця програма забезпечує сканування простору зовнішньою цифровою камерою,
що встановлена на поворотній платформі. Сканування здійснюється по горизонталі та по вертикалі,
кількість точок сканування повинна бути не менше 5 х 5. Тип сканування та оператори циклу, які
забезпечують це сканування, вказані у таблиці 2.
2. Друга програма зчитує та відображає ЦВ з зовнішньої цифрової камери.
